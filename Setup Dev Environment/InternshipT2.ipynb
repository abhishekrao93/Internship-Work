```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
data
```

```python
#We use groupBy to split the data in terms of its rating and get the count based on the ratings

grouped_rating = data.groupby('Y')['Id'].count()
print(grouped_rating)
```

```python
#Draw a bar chart of the "quality rating column" using matplotlib

ratings = ['HQ','LQ EDIT','LQ CLOSE']
plt.xlabel("Quality Rating")
plt.ylabel("Count")
plt.title("Count per Quality Rating")
plt.bar(ratings,grouped_rating)
plt.show()
```
